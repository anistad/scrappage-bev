Years remaining (years before the vehicles will be scrapped): 


```python
years_left = pd.DataFrame({'Pre Euro 0': [3.33, 2.22,1.54], 'Euro 0': [4.10, 2.88, 2.14], 'Euro 1': [5, 3.77, 3.01], 'Euro 2': [6.22,5.12, 4.41]})
years_left.index = ['High', 'Medium', 'Low']
print(years_left)
scenario = 'Medium'

avg_years_left = sum([sum(df_dist_old_g.iloc[:,i]+df_dist_old_d.iloc[:,i])/(sum(df_dist_old_g.sum())+sum(df_dist_old_d.sum()))*years_left.loc[scenario][i] for i in [0,1,2,3]])
```

Mileage used:

```python
mileage_diesel = df_diesel_average.iloc[0:7]
mileage_gasoline = df_gasoline_average.iloc[0:7]
mileage_gasoline
```

Calculating the total kilometers driven in the program period by the old and new vehicles (by emission concept and size):

```python
#Multiplying mileage and years left to obtain the total kilometer driven per cars in the years the program is effective
km_old_g = pd.concat([mileage_gasoline.iloc[:,i]*years_left.loc[scenario][i] for i in [0,1,2,3]],axis=1)

km_old_d = pd.concat([mileage_diesel.iloc[:,i]*years_left.loc[scenario][i] for i in [0,1,2,3]], axis=1)

index_labels = km_old_g.sort_index().index

#Multiplying with distribution of cars to obtain the total kilometer driven for all cars in the program
total_km_old_g = pd.DataFrame(np.multiply(km_old_g.sort_index().to_numpy().T,df_dist_old_g.sort_index().T.to_numpy())).T
total_km_old_g.index = index_labels

total_km_old_g.columns = ['pre euro 0', 'euro 0', 'euro 1', 'euro 2']

total_km_old_d = pd.DataFrame(np.multiply(km_old_d.sort_index().to_numpy().T,df_dist_old_d.sort_index().T.to_numpy())).T
total_km_old_d.index = index_labels
total_km_old_d.columns = ['pre euro 0', 'euro 0', 'euro 1', 'euro 2']
```

```python
#Multiplying the average annual mileage with the average years remaining - yields km in program period per car

km_new_g = pd.concat([mileage_gasoline.loc[:,'new']*avg_years_left, mileage_gasoline.loc[:,'new']*avg_years_left,mileage_gasoline.loc[:,'new']*avg_years_left], axis = 1, keys = ['euro 4', 'euro 5', 'euro 6']).T
km_new_d = pd.concat([mileage_diesel.loc[:,'new']*avg_years_left, mileage_diesel.loc[:,'new']*avg_years_left,mileage_diesel.loc[:,'new']*avg_years_left], axis = 1, keys = ['euro 4', 'euro 5', 'euro 6']).T

#Multiplying with the number of cars 
index_labels = km_new_g.T.sort_index().index
total_km_new_g = np.multiply(km_new_g.T.sort_index(), df_dist_new_g.sort_index()) 
total_km_new_g.index = index_labels

total_km_new_d = np.multiply(km_new_d.T.sort_index(), df_dist_new_d.sort_index()) 
total_km_new_d.index = index_labels
```

Plotting total kilometers driven in the program period for the old and new vehicles: 

```python
plot_old = pd.concat([pd.DataFrame(total_km_old_g.sum(axis = 0)).sum(axis =0), pd.DataFrame(total_km_old_d.sum(axis = 0)).sum(axis = 0)], keys = ['ICEV-p', 'ICEV-d'])
plot_new = pd.concat([pd.DataFrame(total_km_new_g.sum(axis = 0)).sum(axis =0), pd.DataFrame(total_km_new_d.sum(axis = 0)).sum(axis = 0)], keys = ['ICEV-p', 'ICEV-d'])
fig = plt.figure(figsize = [3.5,3.5])
ax = pd.concat([plot_old,plot_new], keys = ['clunkers', 'new']).unstack().unstack(level = 1).plot.bar(stacked = True, legend = None, rot = 0, fontsize = 8)
ax.legend(['ICEV-p', 'ICEV-d'],bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize = 7)
t = ax.yaxis.get_offset_text()
t.set_size(8)
ax.set_ylabel('Total annual kilometers driven', fontsize = 8)
#plt.savefig('tot kilometers cfc program.svg', bbox_inches = 'tight')
```

## Calculation of life-cycle emissions for the BAU and RaR scenarios

In this section the comparasion between the two different scenarios are considered: either BAU or CFC. 
BAU implies that the old cars are driven during the period between early and natural scrapping, while CFC implies that the new vehicles are driven in this period. Additionally, CFC needs to account for the production and disposal emissions from these attributed to the specific period. 
 

Variables: 
- total_km_old dataframe with km driven in the years remaining by the diesel/petrol cars by euro class and size 
- total_km_new dataframe with km driven in the years remaining by the diesel/petrol cars by euro class and size 
- older_cars_lca_agg dataframe with lca score for conventional vehicles for each euro class
#- impact_old_cars stores the LCA impact of the cars in the years remaining (LCA impact/km * tot_km_old)

The LCA results are calculated for a lifetime of 200 000 km. 

```python
older_cars_lca_agg.head()
```

Defining function to calculate results and plot for each county: 

```python
kg_to_mt = 1/(1000*1000*1000)
```

#### Calculate life-cycle impact of program

```python
#Same as above, but for aggregated life cycle phases, i.e. db_lca_agg

size_classes = ['Mini', 'Small', 'Lower medium', 'Medium', 'Large', 'Van', 'SUV']
#list to store dataframes for impact, FU*km driven 
impact_new_cars = list() 
impact_old_cars = list()


for index, row in older_cars_lca_agg.iterrows(): 

    for size in size_classes:
        
        #loop to multiply correctly with km driven for new cars 
        for euro in ['euro 4', 'euro 5', 'euro 6']: 
            if size in str(index[1]) and euro in str(index[0]):
                if 'ICEV-d' in str(index[1]):
                   
                    
                    temp1 = pd.DataFrame(row) #picking the correct row from initial lca dataframe 
                    
                    #multiplying impacts per km by km driven for specific euro class and size 
                    temp1[temp1.select_dtypes(include=['number']).columns] *=total_km_new_d.loc[size,euro] #multiplying with km for size and euro concept 
                    impact_new_cars.append(temp1.T)
                if 'ICEV-p' in str(index[1]): 
                    temp2 = pd.DataFrame(row)
                    temp2[temp2.select_dtypes(include=['number']).columns] *=total_km_new_g.loc[size,euro] #multiplying with km for size and euro concept 
                    impact_new_cars.append(temp2.T)
                
        #loop to multiply correctly with km driven for old cars 
        for euro2 in ['pre euro 0']:
            if size in str(index[1]) and euro2 in str(index[0]): 
                if 'ICEV-d' in str(index[1]):
                   
                    temp1 = pd.DataFrame(row)
                    temp1[temp1.select_dtypes(include=['number']).columns] *=total_km_old_d.loc[size,euro2] 
                    impact_old_cars.append(temp1.T)
                if 'ICEV-p' in str(index[1]): 
                    
                    temp2 = pd.DataFrame(row)
                    temp2[temp2.select_dtypes(include=['number']).columns] *=total_km_old_g.loc[size,euro2] 
                    impact_old_cars.append(temp2.T)
                    

            
        for euro1 in ['euro 0', 'euro 1', 'euro 2']: 
            if size in str(index[1]) and euro1 in str(index[0]) and 'pre' not in str(index[0]): 
                    if 'ICEV-d' in str(index[1]):
                        temp1 = pd.DataFrame(row)
                        temp1[temp1.select_dtypes(include=['number']).columns] *=total_km_old_d.loc[size,euro1] 
                        impact_old_cars.append(temp1.T)
                        
                    if 'ICEV-p' in str(index[1]): 
                        temp2 = pd.DataFrame(row)
                        temp2[temp2.select_dtypes(include=['number']).columns] *=total_km_old_g.loc[size,euro1] 
                        impact_old_cars.append(temp2.T)
                   
                    
#create dataframes with impacts for the total km driven                     
impact_old_cars = pd.concat(impact_old_cars).drop('prod_disp', axis = 1, level=1)
impact_new_cars = pd.concat(impact_new_cars)

     
                
```

#### Plotting impact of program

```python
results_cfc = list() 
results_bau = list()

fig = plt.figure()
changed_color_order = sns.color_palette()[0], sns.color_palette()[2], sns.color_palette()[1], sns.color_palette()[-1], sns.color_palette()[4]
unit = ['Mt CO2 eq.','Mt NMVOC', 'Mt PM10 eq.']
j = 1
cats = ['CC','POF', 'PMF']
for cat in cats :
        
    
    # Including the share of production and disposal which is attributed to the program period
    bau = impact_old_cars[cat].sum(axis = 0)*kg_to_mt
    results_bau.append(bau)
    cfc = impact_new_cars[cat].sum(axis = 0)*kg_to_mt #including the production for new cars 
    results_cfc.append(cfc)
    
    # Plotting result
    ax = fig.add_subplot(1,3,j)
    ax = pd.concat([cfc, bau], keys = ['cfc', 'bau']).unstack()[['fuel', 'direct', 'market for road', 'maintenance, passenger car', 'prod_disp']].plot(kind='bar', stacked=True, ax = ax, legend = None, fontsize = 14, figsize = (10,5), color=changed_color_order)
    ax.set_ylabel(unit[j-1], fontsize = 14)
    ax.set_title(cat, fontsize = 14)
    j+=1

#ax.legend(['fuel', 'direct', 'road', 'maintenance', 'production and disposal'],bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

handles, labels = ax.get_legend_handles_labels()
labels = ['Diesel/petrol production', 'Direct', 'Road', 'Maintenance', 'Production and disposal']
ax.legend(reversed(handles), reversed(labels), loc='upper left', bbox_to_anchor = (1.05,1))


fig.tight_layout()
#fig.savefig('lca results.svg')
```

Calculate the savings in Mt (>0 increased emissions, <0 decreased emissions): 

```python
savings = pd.DataFrame(results_bau).T.sum()-pd.DataFrame(results_cfc).T.sum()
savings.index = cats
savings
```

Defining then a function similar as done for each of the LCIA impact categories to calculate the emissions of the cfc and bau scenarios and calculate the savings: 

```python
#Same as above, but for aggregated life cycle phases, i.e. db_lca_agg

def calc_air_poll(df): #df is the dataframe with emission factors for air pollutants: so pmfp_icev or pofp_icev
    #list to store dataframes for impact, FU*km driven 
    list_air_poll_new = list()
    list_air_poll_old = list()

    for index, row in df.iterrows(): 

        for size in size_classes:
            if size == 'Lower medium': 
                size = 'Lower'

            #loop to multiply correctly with km driven for new cars 
            for euro in ['euro 4', 'euro 5', 'euro 6']: 
                if size in str(index[1]) and euro in str(index[0]):
                    if size == 'Lower': 
                        size = 'Lower medium'
                    if 'ICEV-d' in str(index[1]):
                        #print(size, euro, km_new_d.loc[euro, size]) #size and the correct km driven

                        temp1 = pd.DataFrame(row) #picking the correct row from initial lca dataframe 

                        #multiplying impacts per km by km driven for specific euro class and size 
                        temp1[temp1.select_dtypes(include=['number']).columns] *=total_km_new_d.loc[size,euro] #multiplying with km for size and euro concept 
                        list_air_poll_new.append(temp1.T)
                    if 'ICEV-p' in str(index[1]): 
                        temp2 = pd.DataFrame(row)
                        temp2[temp2.select_dtypes(include=['number']).columns] *=total_km_new_g.loc[size,euro] #multiplying with km for size and euro concept 
                        list_air_poll_new.append(temp2.T)

            for euro2 in ['pre euro 0']:
                if size in str(index[1]) and euro2 in str(index[0]): 
                    if size == 'Lower': 
                        size = 'Lower medium'
                    if 'ICEV-d' in str(index[1]):
            #print(size, euro2, km_old_d.loc[euro2, size]) #size and the correct km driven
                        temp1 = pd.DataFrame(row)
                        temp1[temp1.select_dtypes(include=['number']).columns] *=total_km_old_d.loc[size,euro2] 
                        list_air_poll_old.append(temp1.T)
                    if 'ICEV-p' in str(index[1]): 
                        #print(size, euro2, km_old_g.loc[euro2,size])
                        temp2 = pd.DataFrame(row)
                        temp2[temp2.select_dtypes(include=['number']).columns] *=total_km_old_g.loc[size,euro2] 
                        list_air_poll_old.append(temp2.T)



            for euro1 in ['euro 0', 'euro 1', 'euro 2']: 
                if size in str(index[1]) and euro1 in str(index[0]) and 'pre' not in str(index[0]): 
                    if size == 'Lower': 
                        size = 'Lower medium'
                    if 'ICEV-d' in str(index[1]):
                        temp1 = pd.DataFrame(row)
                        temp1[temp1.select_dtypes(include=['number']).columns] *=total_km_old_d.loc[size,euro1] 
                        list_air_poll_old.append(temp1.T)
                        #print(size, euro1, km_old_d.loc[euro1, size]) #size and the correct km driven
                    if 'ICEV-p' in str(index[1]): 
                        temp2 = pd.DataFrame(row)
                        temp2[temp2.select_dtypes(include=['number']).columns] *=total_km_old_g.loc[size,euro1] 
                        list_air_poll_old.append(temp2.T)
                        #print(size, euro1, km_old_g.loc[euro1,size])
                    
    return pd.concat(list_air_poll_old), pd.concat(list_air_poll_new)
     
                
```

### Plotting air pollution results

```python
nmvoc_new = pd.DataFrame(calc_air_poll(older_cars_POF)[1].drop('NOx', level=2,axis=0).sum(axis=0))
nmvoc_new.columns = ['NMVOC']
air_poll_new = pd.concat([calc_air_poll(older_cars_PMF)[1].groupby(level=2).sum(),nmvoc_new.T])

nmvoc_old = pd.DataFrame(calc_air_poll(older_cars_POF)[0].drop('NOx', level=2,axis=0).sum(axis=0))
nmvoc_old.columns = ['NMVOC']
air_poll_old = pd.concat([calc_air_poll(older_cars_PMF)[0].groupby(level=2).sum(),nmvoc_old.T])
air_poll_old = air_poll_old.drop('prod_disp', axis=1)
```

```python
indirect = air_poll_new.drop('direct', axis=1).sum(axis=1)*kg_to_mt-air_poll_old.drop('direct', axis=1).sum(axis=1)*kg_to_mt

direct = air_poll_new['direct']*kg_to_mt-air_poll_old['direct']*kg_to_mt
sns.set_style("darkgrid")
ax = pd.concat([direct, indirect], axis = 1, keys = ['direct', 'indirect']).T[['PM2.5', 'PM10', 'NOx', 'SO2', 'NH3', 'NMVOC']].T.plot.bar(stacked = True, rot = True, figsize = [10,5], fontsize = 14)
```
